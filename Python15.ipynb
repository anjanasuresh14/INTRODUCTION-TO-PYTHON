```python
import pandas as pd
df=pd.read_csv('SOCR-HeightWeight.csv')
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
h=df['Height(Inches)']
w=df['Weight(Pounds)']
print('Mean Height ',h.mean())
print('Mean Weight ',w.mean())
print('Median Height ',h.median())
print('Median Weight ',w.median())
print('Mode Height ',h.mode())
print('Mode Weight ',w.mode())
plt.xlabel('Weight')
plt.ylabel('Height')
plt.scatter(h,w,color='blue')
```

```python
#weight of the people are in a positive relation with the height
#as height increase, weight also increases proportionally
#it can be used in a textile shop to know about the production
```
